```python
class Kitap:
    def __init__(self, baslik, yazar, yayin_evi, yayin_yili):
        self.baslik = baslik
        self.yazar = yazar
        self.yayin_evi = yayin_evi
        self.yayin_yili = yayin_yili
        
    def bilgileri_göster(self):
        print("n\ Kitap Bilgileri: ")
        print(f"Başlık: {self.baslik}\nYazar: {self.yazar}\nYayın Evi: {self.yayin_evi}\nYayin Yılı: {self.yayin_yili}")

class Uye:
    def __init__(self, ad, soyad, uye_numarası):
        self.ad = ad
        self.soyad = soyad
        self.uye_numarası = uye_numarası
        
    def bilgileri_göster(self):
        print("\n Kitap Bilgileri: ")
        print(f"Ad: {self.ad}\nSoyad: {self.soyad}\nÜye Numarası: {uye_numarası}")

class Kutuphane:
    def __init__(self):
        self.kitaplar = []
        self.uyeler = []
        self.odunc_alinmis_kitaplar = {}
        
    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)
        print(f"{kitap.baslik} kütüphanemizde başarılı bir şekilde eklendi.")
            
    def kitap_sil(self, baslik):
        for kitap in self.kitaplar:
            if kitap.baslik == baslik:
                self.kitaplar.remove(kitap)
                print(f"{kitap.baslik} kütüphanemizden başarılı bir şekilde silindi.")
        print(f"{baslik} baslıklı kitap bulunamadı.")
        
    def uye_ekle(self, uye):
        self.uyeler.append(uye)
        print(f"{uye.ad} {uye.soyad} kütüphanemizin üyeleri arasına başarılı bir şekilde eklendi.")
        
    def uye_sil(self, uye_numarası):
        for uye in self.uyeler:
            if uye.uye_numarası == uye_numarası:
                self.uyeler.remove(uye)
                print(f"{uye.uye_numarası} kütüphanemizden başarılı bir şekilde silindi.")
                return
        print(f"{uye_numarası} numara sahip üye bulunamadı.")
            
        
    def odunc_almak(self, kitap, uye):
        if kitap in self.kitaplar and uye in self.uyeler:
            if kitap not in self.oduc_alınmış_kitaplar:
                self.oduc_alınmış_kitaplar[kitap] = uye
                print(f"{kitap.baslik} adli kitap {uye.ad} {uye.soyad} kişisi tarafından başarılı bir şekilde ödünç alındı")
            else:
                print(f"{kitap.baslik} daha önce başka birisi tarafından ödünç alınmıştır.")
    def geri_vermek(self, kitap):
        if kitap in self.oduc_alınmış_kitaplar:
            uye = self.oduc_alınmış_kitaplar.pop(kitap)
            print(f"{kitap.baslik} adlı kitap {uye.ad} {uye.soyad} kişisi tarafından geri verildi.")
        else:
            print(f"{kitap.baslik} adlı kitap şuan ödünçte değildir.")
            
    def kitapları_göster(self):
        if len(self.kitaplar) == 0:
            print("Henüz kütüphanemize eklenmiş bir kitap yoktur.")
        else:
            print("\nKitaplar: ")
            for kitap in self.kitaplar:
                kitap.bilgileri_göster()
      
    
    def uyeleri_göster(self):
        if len(self.uyeler) == 0:
            print("Henüz kütüphanemizde eklenmiş bir üye yoktur.")
        else:
            print("\nÜyeler: ")
            for uye in self.uyeler:
                uye.bilgileri_göster()
        
    
kutuphane = Kutuphane()
while True:
    print(7*".*.", "Kütüphane Yönetim Sistemi Menüsü", 7*".*.")
    print("""
            1.Kitap Ekle
            2.Kitap Sil
            3.Üye Ekle
            4.Üye Sil
            5.Kitap Ödünç Al
            6.Kitap Geri Ver
            7.Kitapları Göster
            8.Üyeleri Göster
            9. Çıkış (q)  
    """)
    islem = input("Yapacagınız işlemin numarasını giriniz: ")
    if islem == "1":
        baslik = input("Kitap başlıgı: ")
        yazar = input("Yazar: ")
        yayin_evi = input("Yayin_evi: ")
        yayin_yili = input("Yayın Yılı: ")
        kitap = Kitap(baslik, yazar, yayin_evi, yayin_yili)
        kutuphane.kitap_ekle(kitap)
        
    elif islem == "2":
        baslik =input("Silmek istediğiniz kitabın adını giriniz: ")
        kutuphane.kitap_sil(baslik)
        
    elif islem == "3":
        ad = input("Üye adı: ")
        soyad = input("Üye soyadı: ")
        uye_numarası = input("Üye numarası: ")
        uye = Uye(ad, soyad, uye_numarası)
        kutuphane.uye_ekle(uye)
        
    elif islem =="4":
        uye_numarası = input("silmek istediğiniz üyenin numarasını giriniz: ")
        kutuphane.uye_sil(uye_numarası)
        
    elif islem =="5":
        baslik = input("Kitabın adını giriniz: ")wa
        uye_numarası = input("Üye numarasını giriniz: ")
        kitap = next((k for k in kutuphane.kitaplar if k.baslik == baslik), None)
        uye = next((u for u in kutuphane.uyeler if u.uye_numarası == uye_numarası), None)
        if kitap and uye:
            kutuphane.odunc_almak(kitap, uye)
        else:
            print("Yanlış bir kitap veya üye girişi yaptınız.")
            
            
    elif islem == "6":
        baslik = input("Kitabın adını giriniz: ")
        kitap = next((k for k in kutuphane.kitaplar if k.baslik == baslik), None)
        if kitap:
            kutuphane.geri_vermek(kitap)
        else:
            print("Yanlış bir kitap girişi yaptınız.")
            
    elif islem == "7":
        kutuphane.kitapları_göster()
    
    elif islem == "8":
        kutuphane.uyeleri_göster()
        
    elif islem =="9" or islem == "q" or islem == "Q":
        print("Uygulamadan çıkış yaptınız.")
        break
    else:
        print("Yanlış bir işlem seçimi yaptınız.:(  Lütfen yapmak istediğiniz işlemin sadece numarasını giriniz. :)")
        
```
